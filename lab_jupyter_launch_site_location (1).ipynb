Let's first create a `MarkerCluster` object

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Create a list of the proximities of interest
proximity = {
    'Kennedy Space Center': [28.5721, -80.648],
    'Cape Canaveral': [28.4058, -80.6214],
    'Vandenberg Space Launch Complex': [34.6328, -120.6107],
    'SpaceX South Texas Launch Site': [25.9970, -97.1561]
}

# For each launch site, use haversine formula to calculate the distance between the site to the proximity
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

for site in proximity.keys():
    distance = haversine(nasa_coordinate[0], nasa_coordinate[1], proximity[site][0], proximity[site][1])
    print(f"Distance from NASA Johnson Space Center to {site} is {distance} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )


# Adicione um marcador no ponto da costa mais prxima
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Adicione o marcador ao mapa
distance_marker.add_to(site_map)

# Exiba o mapa
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Coordenadas do local de lanamento (exemplo)
launch_site_lat = 28.56367
launch_site_lon = -80.57163

# Coordenadas das proximidades (exemplos)
railway_lat = 28.5721
railway_lon = -80.5853

highway_lat = 28.5634
highway_lon = -80.5707

coastline_lat = 28.56367
coastline_lon = -80.57163

city_lat = 28.3922
city_lon = -80.6077

# Calcula as distncias at as proximidades
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

# Adicione um marcador no local de lanamento
launch_marker = folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup='Launch Site',
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Adicione marcadores nas proximidades
proximities = [
    (railway_lat, railway_lon, distance_railway, 'Railway'),
    (highway_lat, highway_lon, distance_highway, 'Highway'),
    (coastline_lat, coastline_lon, distance_coastline, 'Coastline'),
    (city_lat, city_lon, distance_city, 'City')
]

for lat, lon, distance, name in proximities:
    folium.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
        ),
        popup=name
    ).add_to(site_map)
    
    # Adicione uma linha entre o local de lanamento e a proximidade
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [lat, lon]],
        color='red'
    ).add_to(site_map)

# Exiba o mapa
site_map
```
